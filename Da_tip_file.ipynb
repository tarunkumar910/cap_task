```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('./archive/tips.csv')
df.head()
plt.scatter('total_bill','tip',data=df,color='green',marker='*',label='tip')
plt.bar(x=df.day.value_counts().index,height=df.day.value_counts().values)
no1=df['total_bill'][:31]
no2=df['tip'][51:78]
# Create subplots with various plot types
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
# 1. Line plot
axes[0, 0].plot(no1.values[:27], label='total_bill', marker='o')
axes[0, 0].plot(no2.values, label='tip', marker='s')
axes[0, 0].set_title('Line Plot')
axes[0, 0].legend()
axes[0, 0].set_xlabel('Index')
axes[0, 0].set_ylabel('Value')
# 2. Bar plot
x = np.arange(min(len(no1), len(no2)))
width = 0.35
axes[0, 1].bar(x - width/2, no1.values[:27], width, label='total_bill')
axes[0, 1].bar(x + width/2, no2.values, width, label='tip')
axes[0, 1].set_title('Bar Plot')
axes[0, 1].legend()
axes[0, 1].set_xlabel('Index')
axes[0, 1].set_ylabel('Value')
# 3. Scatter plot
axes[0, 2].scatter(range(len(no1)), no1.values, label='total_bill', alpha=0.7)
axes[0, 2].scatter(range(len(no2)), no2.values, label='tip', alpha=0.7)
axes[0, 2].set_title('Scatter Plot')
axes[0, 2].legend()
axes[0, 2].set_xlabel('Index')
axes[0, 2].set_ylabel('Value')
# 4. Histogram
axes[1, 0].hist(no1.values, bins=10, alpha=0.5, label='total_bill')
axes[1, 0].hist(no2.values, bins=10, alpha=0.5, label='tip')
axes[1, 0].set_title('Histogram')
axes[1, 0].legend()
axes[1, 0].set_xlabel('Value')
axes[1, 0].set_ylabel('Frequency')
# 5. Boxplot
axes[1, 1].boxplot([no1.values[:27], no2.values], labels=['total_bill', 'tip'])
axes[1, 1].set_title('Boxplot')
axes[1, 1].set_ylabel('Value')
# 6. Area plot
axes[1, 2].fill_between(range(len(no1)), no1.values[:27], alpha=0.5, label='total_bill')
axes[1, 2].fill_between(range(len(no2)), no2.values, alpha=0.5, label='tip')
axes[1, 2].set_title('Area Plot')
axes[1, 2].legend()
axes[1, 2].set_xlabel('Index')
axes[1, 2].set_ylabel('Value')
plt.tight_layout()
plt.show()
```
