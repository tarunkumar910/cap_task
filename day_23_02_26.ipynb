```python
class twelve(tenth):
    def __init__(self,name,age,marks,stream):
        super().__init__(name,age,marks)
        
        self.stream=stream
    
    def display(self):
        super().display()
        print(self.stream)
```

```python
# person class common for all


class person:
    def __init__(self,name,age,gender):
        self.name =name
        self.age=age
        self.gender=gender
    def display(self):
        print("Basic deatials: ",self.name,self.age,self.gender)
```

```python
# student class inheriting person class

class student(person):
    def __init__(self,name,age,gender,standard):
        super().__init__(name,age,gender)
        self.standard=standard
    

    def display(self):
        super().display()
        print("Standard: ",self.standard)



# create class of teacher inheriting person class

class teacher(person):
    def __init__(self,name,age,gender,subject_teacher,salary):
        super().__init__(name,age,gender)
        self.subject_teacher=subject_teacher
        self.salary=salary

    def display(self):
        super().display()
        print(self.subject_teacher,self.salary)
```

```python
# creating the class for academic for the student


class academic:
    def __init__(self,marks,feedback):
        
        self.marks=marks
        self.feedback=feedback
    def display(self):
        
        print("Academic deatails: ",self.marks,self.feedback)


# creating the class for the sports for the student

class sports:
    def __init__(self,sport_feedback):
        
        self.sport_feedback=sport_feedback

    def display(self):
        
        print("Sport feedback: ",self.sport_feedback)


        
```

```python
# now creating all rounder class for the student that will inherited the properties of both academic and sports class


class all_rounder(student,academic,sports):
    def __init__(self, name, age, gender, standard, marks, feedback, sport_feedback):
        student.__init__(self,name,age,gender,standard)
        academic.__init__(self,marks, feedback)
        sports.__init__(self,sport_feedback)
        

    def display(self):
        student.display(self)
        academic.display(self)
        sports.display(self)

    def check_student_area_of_interset(self):
        if self.marks>=75 and self.sport_feedback.lower()=="excellent":
            print("the student is an all rounder")
        elif self.marks>=75 and self.sport_feedback.lower()!="excellent":
            print("the student is an academician")
        elif self.marks<75 and self.sport_feedback.lower()=="excellent":
            print("the student is a sportsperson")
        else:
            print("the student is a normal student")

        
    
```

```python
obj3=all_rounder("tarun",23,"M",12,90,"good","excellent")
```

```python
obj3.display()
```

```python
obj3.check_student_area_of_interset()
```
