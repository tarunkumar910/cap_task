```python
# person class common for all


class person:
    def __init__(self,name,age,gender):
        self.name =name
        self.age=age
        self.gender=gender
    def display(self):
        print(self.name,self.age,self.gender)
```

```python
# student class inheriting person class

class student(person):
    def __init__(self,name,age,gender,standard):
        super().__init__(name,age,gender)
        self.standard=standard
    

    def display(self):
        super().display()
        print(self.standard)



# create class of teacher inheriting person class

class teacher(person):
    def __init__(self,name,age,gender,subject_teacher,salary):
        super().__init__(name,age,gender)
        self.subject_teacher=subject_teacher
        self.salary=salary

    def display(self):
        super().display()
        print(self.subject_teacher,self.salary)
```

```python
# creating the class for academic for the student


class academic(student):
    def __init__(self,name,age,gender,standard,marks,feedback):
        super().__init__(name,age,gender,standard)
        self.marks=marks
        self.feedback=feedback
    def display(self):
        super().display()
        print(self.marks,self.feedback)


# creating the class for the sports for the student

class sports(student):
    def __init__(self,name,age,gender,standard,sport_feedback):
        super().__init__(name,age,gender,standard)
        self.sport_feedback=sport_feedback

    def display(self):
        super().display()
        print(self.sport_feedback)


        
```

```python
# now creating all rounder class for the student that will inherited the properties of both academic and sports class


class all_rounder(academic,sports):
    def __init__(self, name, age, gender, standard, marks, feedback, sport_feedback):
        academic.__init__( name, age, gender, standard, marks, feedback)
        self.sport_feedback = sport_feedback

    def display(self):
        academic.display(self)
        print(self.sport_feedback)
    
```

```python
obj3=all_rounder("tarun",23,"M",12,90,"good","excellent")
```
