```python
# the base class vechile

class vechile:
    def __init__(self,id,brand):
        self.id=id
        self.brand=brand

    def display(self):
        print("vechile deatails: ",self.id,self.brand)

    def calculate_rent():
        pass
```

```python
# creating car class inthering the propertis of vechile class

class car(vechile):
    def __init__(self,id,brand,model):
        super().__init__(id,brand)
        self.model=model
    def display(self):
        super().display()
        print("car deatails: ",self.model)

    

    def calculate_rent(self,days):
        return "per day rent",days*1500
    



# creting bike class 

class bike(vechile):
    def __init__(self,id,brand,model):
        super().__init__(id,brand)
        self.model=model
    def display(self):
        super().display()
        print("bike deatails: ",self.model)

    

    def calculate_rent(self,hr):
        return "per hour rent",hr*100
    


# creating the truck deatils

class truck(vechile):
    def __init__(self,id,brand,model):
        super().__init__(id,brand)
        self.model=model
    def display(self):
        super().display()
        print("truck deatails: ",self.model)

    

    def calculate_rent(self,hr):
        return "per km rent",hr*150
```

```python
car1=car(123,"BMW","X5")
car1.display()
```

```python
print(car1.calculate_rent(5))
```
