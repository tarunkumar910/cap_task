```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(1,50,10)
y=np.random.randint(1,50,10)
print(x,y,sep='\n')
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color='red',linestyle='dotted')
plt.show()
```

```python
plt.title(("random chart"))
plt.plot(x,y,color='red',linestyle='dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.plot(x,y,color='red',linestyle='dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.xticks()
plt.yticks()
plt.plot(x,y,color='red',linestyle='dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='red',linestyle='dotted')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='red',linestyle='-',alpha=0.9,
         marker='o')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='red',linestyle='-',alpha=0.9,
         marker='*')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='blue',linestyle='-',alpha=0.9,
         marker='*',markerfacecolor='y')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='blue',linestyle='-',alpha=0.9,
         marker='*',markerfacecolor='y' ,markersize=10)
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='blue',linestyle='-',alpha=0.9,
         marker='*',markerfacecolor='y' ,markersize=10,markeredgecolor='black')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color='blue',linestyle='-',alpha=0.9,
         marker='*',markerfacecolor='y' ,markersize=10,linewidth=3)
plt.show()
```

```python
x=np.arange(1,11)
y=np.sin(x)
z=np.cos(x)
print(x,y,z,sep='\n\n')
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,label='SINQ')
```

```python
plt.plot(x,y,label='SINQ')
plt.plot(x,z,label='COSQ')
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,label='SINQ',color='red')
plt.plot(x,z,label='COSQ',color='black')
plt.title("SINQ & COSQ")
plt.xlabel=('Data points')
plt.ylabel=("Associate value")
plt.xticks()
plt.yticks()
plt.legend()
plt.show()
```
