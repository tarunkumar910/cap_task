```python
def count(a):
    odd=0
    even=0

    while a !=0:
        digit = a%10
        if digit%2 ==0:
            even=even+1
        else:
            odd=odd+1
        a=a//10

    return even , odd

count(12345)
```

```python
def palidrom(num):
    temp =num
    rev=0
    while temp !=0:
        rev =rev*10+temp%10
        temp =temp//10

    if rev == num:
        return "pallidrom"
    return "not pallidrom"

palidrom(23432)
```
