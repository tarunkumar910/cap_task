```python
# exception handiling
def safe_drive():
    try:
        a=int(input("enter a number"))
        b=int(input("enter the number"))
        c=a/b
        print(c)
    except ValueError:
        print("invalid value")
        safe_drive()
    except ZeroDivisionError:
        print("zero divisi error")
        safe_drive()
```

```python
safe_drive()
```

```python
class AgeError(Exception):
    pass


age=int(input("enter the number"))
if age<18:
    raise AgeError("age is below 18")
else:
    print("ok")
```
