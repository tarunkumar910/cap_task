```python
from abc import ABC, abstractmethod

class BankAccount(ABC):

    def __init__ (self, CustomerName, AccountNumber, DOB):
        self.__CustomerName=CustomerName
        self.__AccountNumber=AccountNumber
        self.__DOB=DOB

    
    @abstractmethod
    def deposite(self,amount):
        pass

    @abstractmethod
    def withdraw(self,amount):
        pass

    def display(self):
        print("Customer Name: ",self.__CustomerName)
        print("Account Number: ",self.__AccountNumber)
        print("Date of Birth: ",self.__DOB)



class Savin_BankAccount(BankAccount):
    
    def __init__(self,CustomerName, AccountNumber,DOB):
        super().__init__(CustomerName,AccountNumber,DOB)
        self.__Balance=1000


    def deposite(self, amount):
        if amount>0:
            self.__Balance+=amount
            print("Current Balance: ",self.__Balance)
        
        else:
            print("Invalid Amount")

    def withdraw(self, amount):
        if amount >0 and amount<=self.__Balance:
            if self.__Balance-amount>=1000:
                self.__Balance-=amount
                print("Current Balance: ",self.__Balance)
            else:
                print("Minium Balance should be 1000")

        elif amount>self.__Balance:
            print("Insufficient Balance")
        else:
            print("Invalid Amount")

    def display(self):
        super().display()
        print("Balance: ",self.__Balance)
```
