```python
hello=[]
with open("data.txt","r") as f:
    hello=f.read().split(" ")
    print(len(hello))

```

```python
# creating set 
s1=set()
for char in hello:
    s1.add(char)


print(len(s1))
```

```python
dict1={}

for char in hello:
    if char in dict1:
      dict1[char]+=1
    else:
       dict1[char]=1


```

```python
import json
```

```python
with open("json_data.json","w+") as f:
    json.dump(dict1,f)
    
```

```python
with open("json_data.json","r") as f:
    print(json.load(f))
```
